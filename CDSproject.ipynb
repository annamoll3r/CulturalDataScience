```r
data <- read.csv("CulturalDataScience/Project/Students_from_DK_going_abroad.csv")
```

```r
#removes the X in front of the year
names(data) <- gsub("^X", "", names(data))
```

```r
head(data)
```

```r
#Loading data
distance <- read.csv("CulturalDataScience/Project/Distance_to_DK.csv")
english <- read.csv("CulturalDataScience/Project/English1.csv")
temperature <- read.csv("CulturalDataScience/Project/Temperature.csv")
tourism <- read.csv("CulturalDataScience/Project/Tourism.csv")
cost_of_living <- read.csv("CulturalDataScience/Project/Cost_of_living.csv")
prestige <- read.csv("CulturalDataScience/Project/Prestige.csv")
```

```r
head(data)
```

```r
#makes the 2021 and 2022-columns characters like the rest
data$'2021' <- as.character(data$'2021')
data$'2022' <- as.character(data$'2022')

#Converts the distance-column to numeric
distance$'Distance..km.' <- as.numeric(distance$'Distance..km.')
```

```r
#Pivots the dataframe into long format
data_long <- pivot_longer(
  data,
  cols = `2015`:`2023`,  # Specify the range of year columns
  names_to = "Year",     # New column for years
  values_to = "Danish_exchange_students"    # New column for values
)
```

```r
#Removes the commas in the numbers
#Converts the distance-column to numeric

data_long$Danish_exchange_students <- gsub(',', '', data_long$Danish_exchange_students)
distance$'Distance..km.' <- as.numeric(distance$'Distance..km.')
```

```r
#Adds all the parameters, making one super-dataframe
join1 <- left_join(data_long, distance, by = "Country")
join2 <- left_join(join1, english, by = "Country")
join3 <- left_join(join2, temperature, by = "Country")
join4 <- left_join(join3, tourism, by = "Country")
join5 <- left_join(join4, cost_of_living, by = "Country")
final_df <- left_join(join5, prestige, by = "Country")
```

```r
#Finds possible mismatches between the two dataframes
setdiff(data_long$Country, distance$Country)
setdiff(distance$Country, data_long$Country)
```

```r
#Fixes the mismatches (excludes whitespaces)
data_long <- data_long %>% mutate(Country = trimws(Country))
distance <- distance %>% mutate(Country = trimws(Country))

#Remember to run the left_join functions again after fixing this
```

```r
head(final_df)
```

```r
#Converts blank and NA to 0 in the Danish_exchange_students-column

final_df$Danish_exchange_students <- sub("^$", "0", final_df$Danish_exchange_students)
final_df$Danish_exchange_students[is.na(final_df$Danish_exchange_students)] <- 0
```

```r
head(final_df)
```

```r
#Filters for chosen years (2018 and 2023)

filtered_df <- final_df %>%
  filter(Year %in% c(2018, 2023))

#makes the Danish_exchange_students-column numeric
filtered_df$Danish_exchange_students  <- as.numeric(filtered_df$Danish_exchange_students)
```

```r
head(filtered_df)
```

```r
#Linear model regression for number of exchange students and year

year_model <- lm(Danish_exchange_students ~ Year, filtered_df)
summary(year_model)
```

```r
#Filters for new years (2018 and 2022)

filtered_df2 <- final_df %>%
  filter(Year %in% c(2018, 2022))
```

```r
#Linear model regression number of exchange students and year
year_model2 <- lm(Danish_exchange_students ~ Year, filtered_df2)
summary(year_model2) 
```

```r
#Adds the prestige-predictor to the model
year_prestige_model <- lm(Danish_exchange_students ~ Year + Top.1000, filtered_df2)
summary(year_prestige_model)
```

```r
#Adds the temperature-parameter to the model
year_prestige_temp_model <- lm(Danish_exchange_students ~ Year + Top.1000 + Temperature...C.., filtered_df2)
summary(year_prestige_temp_model)
```

```r
#converts temperature-column to numeric
filtered_df2$'Temperature...C..' <- as.numeric(filtered_df2$'Temperature...C..')
```

```r
year_prestige_temp_model <- lm(Danish_exchange_students ~ Year + Top.1000 + Temperature...C.., filtered_df2)
summary(year_prestige_temp_model)
```

```r
#adds the distance-predictor to the model
year_prestige_temp_dist_model <- lm(Danish_exchange_students ~ Year + Top.1000 + Temperature...C.. + Distance..km., filtered_df2)
summary(year_prestige_temp_dist_model)
```
