## Cutting a numeric variable into categories

```python
#Age distribution of people at social gathering
age = [60,58,24,26,34,42,31,30,33,2,9]
import pandas as pd
df = pd.DataFrame(
    {'age': np.array([60,58,24,26,34,42,31,30,33,2,9])
    })

df
```

```python
#Group these numbers into categories, e.g. 0-20 (young), 21-40 (adult), 41- (older)
#pandas use cut-method to do this
#cut function requires you to specify edge-points


df['categories'] = pd.cut(x = df['age'], bins = [0,20,40,60], labels = ['young', 'adult', 'older'])
df
```

```python
#Three important argument to cut()-function
#x (Variable that needs to be categorized)
#bins (either a vector or a number indicating how many categories you want)
#labels (name of categories)

#Python can make the decisions of e.g. labels and bins

df['categories'] = pd.cut(x = df['age'], bins = 3)
df
```

```python
#Extract information on Python's decisions
pd.crosstab(index = df["categories"], columns = "count")
```

```python
#Python will try to make the bins roughly the same size
#How to make different sizes but with same number of people assigned to them
#Use panda-percentile-function

df['categories'] = pd.qcut(x = df['age'], q = [0, .33, .66, 1] )
df
```

```python
#Measuring RT too
df = pd.read_csv("https://raw.githubusercontent.com/ethanweed/pythonbook/main/Data/drugs.csv")
df.head()
```

```python
#Get wide-to-low for this dataframe
df_long = pd.wide_to_long(df, stubnames = ['WMC','RT'], 
                          i=['id','gender'], 
                          j='drug', 
                          sep = '_', 
                          suffix = '.+')
df_long.head(10)
# Stubnames: columns need to have some kind of groupings for this command to work!
# i: id-variables. Consistent.
# j: name of column where label will follow the prefix
# sep: indicates if some character seperates prefix and label, here '_'
# suffix: expects labels to be numbers, the '+' deletes this assumption
```

```python
#Getting another format (I liked the one before better), better if we want to do further analysis
df_long = df_long.reset_index()
df_long.head(15)
```

```python
#From long to wide format (other way around)
df_wide = pd.pivot(df_long, index=['id'], columns='drug', values=['gender', 'WMC', 'RT'])
df_wide
```

```python
df_WMC = df_wide['WMC']
df_WMC
```

```python
#Get column names
df_WMC.columns = ['WMC_' + col for col in df_WMC.columns]
df_WMC 
#what is col = column
```

```python
#The same for responsetime...
df_RT = df_wide['RT']
df_RT.columns = ['RT_' + col for col in df_RT.columns]
df_RT
```

```python
#The gender variable - only need one of these
df_gender = df_wide['gender']
df_gender
```

```python
#do this to get rid of 2 and rename to gender
df_gender = pd.DataFrame(df_wide['gender']['alcohol']) #alkohol kommer til at hende gender
df_gender.columns = ['gender'] #medtager kun den kolonne
df_gender
```

```python
#Bring them back together using .join()
df_wide = df_gender.join(df_WMC, on = 'id', how = 'left')
df_wide
```

```python
df_wide = df_wide.join(df_RT, on = 'id', how = 'left')
df_wide
```
